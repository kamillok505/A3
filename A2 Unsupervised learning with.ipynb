```python
from sklearn.manifold import TSNE  # Importing t-SNE from the scikit-learn library
import matplotlib.pyplot as plt  # Importing the module for plotting graphs from the matplotlib library
import pandas as pd  # Importing the pandas library for working with data

# Loading data from the file
your_synthetic_data = pd.read_csv('A3-data.txt', delimiter=',')  

# Renaming columns to English names and removing the extra 'class' column
your_synthetic_data.columns = ['x', 'y', 'z', 't', 'class']

# Converting categorical values to numerical using One-Hot Encoding
your_synthetic_data = pd.get_dummies(your_synthetic_data, columns=['class'])

# Creating a column 'Class' indicating the class labels
your_synthetic_data['Class'] = your_synthetic_data[['class_1', 'class_2', 'class_3']].idxmax(axis=1).str[-1].astype(int)

# Separating features and the target variable
y = your_synthetic_data['Class']  # Target variable (Class)
X = your_synthetic_data.drop(['class_1', 'class_2', 'class_3', 'Class'], axis=1)  # Features

# Applying t-SNE with different perplexity values
perplexities = [5, 10, 15, 20]  # Experimenting with different perplexity values
for perplexity in perplexities:
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)  # Creating a t-SNE instance
    X_embedded = tsne.fit_transform(X)  # Transforming the data

    # Visualizing t-SNE projection
    plt.figure(figsize=(8, 6))
    plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='viridis', marker='o')
    plt.title(f't-SNE Projection - A3-data  (Perplexity={perplexity})')  # Title of the plot
    plt.xlabel('Component 1')  # x-axis label
    plt.ylabel('Component 2')  # y-axis label
    plt.colorbar(label='Class')  # Adding a color bar indicating classes
    plt.show()  # Displaying the plot
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score, adjusted_rand_score

# Loading data from the file
your_synthetic_data = pd.read_csv('A3-data.txt', delimiter=',')  

# Renaming columns to English names and removing the extra 'class' column
your_synthetic_data.columns = ['x', 'y', 'z', 't', 'class']

# Converting categorical values to numerical using One-Hot Encoding
your_synthetic_data = pd.get_dummies(your_synthetic_data, columns=['class'])

# Creating a column 'Class' indicating the class labels
your_synthetic_data['Class'] = your_synthetic_data[['class_1', 'class_2', 'class_3']].idxmax(axis=1).str[-1].astype(int)

# Separating features and the target variable
y = your_synthetic_data['Class']  # Target variable (Class)
X = your_synthetic_data.drop(['class_1', 'class_2', 'class_3', 'Class'], axis=1)  # Features

# Real number of classes (you need to know this value)
real_classes = 5

# Applying the k-means method for different k values
k_values = range(2, real_classes + 1)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)

    # Reducing dimensionality using PCA
    pca = PCA(n_components=2)
    reduced_features = pca.fit_transform(X)

    # Comparing the obtained clusters with the real classes (if available)
    accuracy = accuracy_score(y, cluster_labels)
    adjusted_rand = adjusted_rand_score(y, cluster_labels)
    
    print(f"Accuracy for k={k}: {accuracy:.4f}")
    print(f"Adjusted Rand Index for k={k}: {adjusted_rand:.4f}")

    # Visualizing the scatter plot with points colored according to assigned clusters
    plt.figure(figsize=(8, 6))
    plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=cluster_labels, cmap='viridis')
    plt.title(f'k-Means Clustering - A3-data (k={k})')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.colorbar(label='Cluster')
    plt.show()

    # Using a confusion matrix to evaluate the clustering performance
    confusion = confusion_matrix(y, cluster_labels)
    print(f"Confusion Matrix for k={k}:\n{confusion}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram

# Loading data from the file 'A3-data.txt'
your_synthetic_data = pd.read_csv('A3-data.txt', delimiter=',')

# Renaming columns to English names and removing the extra 'class' column
your_synthetic_data.columns = ['x', 'y', 'z', 't', 'class']

# Converting categorical values to numerical using One-Hot Encoding
your_synthetic_data = pd.get_dummies(your_synthetic_data, columns=['class'])

# Creating a column 'Class' indicating the class labels
your_synthetic_data['Class'] = your_synthetic_data[['class_1', 'class_2', 'class_3']].idxmax(axis=1).str[-1].astype(int)

# Converting boolean columns to integer format for numerical analysis
bool_columns = ['class_1', 'class_2', 'class_3', 'class_4', 'class_5', 'class_6']
your_synthetic_data[bool_columns] = your_synthetic_data[bool_columns].astype(int)

# Selecting only numerical columns
numerical_data = your_synthetic_data.drop('Class', axis=1)

# Computing the Euclidean distance matrix for numerical data
distance_matrix = pdist(numerical_data, metric='euclidean')
distance_matrix = squareform(distance_matrix)  # Converting to a square matrix

# Agglomerative Hierarchical Clustering (AHC) with UPGMA linkage
upgma = linkage(distance_matrix, method='average')

# Plotting the UPGMA dendrogram with modifications for better readability
plt.figure(figsize=(68, 60))  # Enlarged plot size
dendrogram(upgma)
plt.title('UPGMA Dendrogram  - A3-data', fontsize= 40)  # Increased font size of the title
plt.ylabel('Distance', fontsize=40)  # Increased font size of the Y-axis label
plt.xticks(rotation=90, fontsize=10)  # Rotating and enlarging the X-axis labels
plt.yticks(fontsize=25)  # Increased font size of the Y-axis ticks
plt.show()  # Display the dendrogram
```

```python
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Loading data from the file
your_synthetic_data = pd.read_csv('A3-data.txt', delimiter=',')

# Renaming columns to English names and removing the extra 'class' column
your_synthetic_data.columns = ['x', 'y', 'z', 't', 'class']

# Converting categorical values to numerical using One-Hot Encoding
your_synthetic_data = pd.get_dummies(your_synthetic_data, columns=['class'])

# Creating a column 'Class' indicating the class labels
your_synthetic_data['Class'] = your_synthetic_data[['class_1', 'class_2', 'class_3']].idxmax(axis=1).str[-1].astype(int)

# Converting boolean columns to integer format for numerical analysis
bool_columns = ['class_1', 'class_2', 'class_3', 'class_4', 'class_5', 'class_6']
your_synthetic_data[bool_columns] = your_synthetic_data[bool_columns].astype(int)

# Extracting numerical features
numerical_features = your_synthetic_data[['x', 'y', 'z', 't']]

# Converting to NumPy array
X = numerical_features.values

# SOM Initialization and Training
som_width = 10
som_height = 10
input_len = X.shape[1]
som = MiniSom(som_width, som_height, input_len, sigma=0.3, learning_rate=0.5)
som.train_random(X, 100)

# Creating the U-Matrix
u_matrix = som.distance_map()

# Visualization of the U-Matrix
plt.figure(figsize=(8, 8))
plt.pcolor(u_matrix, cmap='bone_r')
plt.colorbar()
plt.title('SOM U-Matrix Visualization - A3-data')
plt.show()

```

```python
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Loading data from the CSV file
your_data = pd.read_csv('6 class csv.csv')

# Converting strings to numbers if necessary
data_array = your_data.apply(pd.to_numeric, errors='coerce').fillna(0).values

# Determining the SOM size
map_width = 10
map_height = 10
input_len = data_array.shape[1]  # Determining the dimensionality of input data

# Initializing the SOM
som = MiniSom(map_width, map_height, input_len, sigma=0.3, learning_rate=0.5)

# Training the SOM on the data
som.train_random(data_array, 100)  # 100 training iterations (can be changed)

# Creating the U-Matrix
u_matrix = som.distance_map()

# Visualization of the U-Matrix
plt.figure(figsize=(10, 10))  # Increasing the figure size

# Plotting the U-Matrix
plt.pcolor(u_matrix.T, cmap='bone_r', alpha=0.7, edgecolors='w', linewidths=0.1)
plt.colorbar()
plt.title('SOM U-Matrix Visualization - 6 class csv.csv')  # Graph title
plt.show()  # Display the graph
```
